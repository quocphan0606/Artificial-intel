```python
import random
import numpy as np
class Dense:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
```

```python
#cau 4

inputs = [1, 2, 3]

weights = [
    [0.2, 0.8, -0.5],
    [0.5, -0.91, 0.26],
    [-0.26, -0.27, 0.17]
]

bias = [2, 3, 0.5]

outputs = []

for neuron_weights, neuron_bias in zip(weights, bias):
    neuron_output = sum(i*w for i, w in zip(inputs, neuron_weights)) + neuron_bias
    outputs.append(neuron_output)

print("Outputs:", outputs)
```

```python
# cau 6 
import numpy as np

inputs = np.array([1, 2, 3, 2.5])

# Layer 1
weights1 = np.array([
    [0.2, 0.8, -0.5, 1],
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
])

bias1 = np.array([2, 3, 0.5])

layer1_output = np.dot(weights1, inputs) + bias1

# Layer 2
weights2 = np.array([
    [0.1, -0.14, 0.5],
    [-0.5, 0.12, -0.33],
    [-0.44, 0.73, -0.13]
])

bias2 = np.array([-1, 2, -0.5])

layer2_output = np.dot(weights2, layer1_output) + bias2

print("Final output:", layer2_output)
```

```python
#cau 8 
import numpy as np

class Dense_Layer:
    def __init__(self, n_inputs, n_neurons):
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))

    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
```

```python
#cau 9
layer1 = Dense_Layer(4, 3)

inputs = np.array([
    [1, 2, 3, 2.5]
])

layer1.forward(inputs)

print(layer1.output)
```

```python
# cau 10 
layer1 = Dense_Layer(4, 3)
layer2 = Dense_Layer(3, 3)

inputs = np.array([
    [1, 2, 3, 2.5]
])

layer1.forward(inputs)
layer2.forward(layer1.output)

print("Final output:", layer2.output)
```
